# Etapa 4: Configuração da base de dados de treinamento

# Etapa 5: Construção e treinamento do modelo

* input_shape = Entrada da Rede Neural (passado como array)
* units = saída da rede neural
* keras = tensorflow

```
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
model.summary()
```



*   Total de Parâmetros = Param 2


```
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
```

* Adam = Algoritmo muito utilizado.
* Parametro (0.1) taxa de aprendizado:
* mean_squared_error = Diferença entre o que a RN previu x o valor real

```
epochs_hist = model.fit(X_train, y_train, epochs = 800)
```

# Etapa 5: Avaliação do modelo

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.legend(['Training loss']);
```
